### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

# Calcular ingresos totales por tienda
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

# Mostrar gráfico de ingresos
plt.figure(figsize=(8, 6))
plt.bar(ingresos.keys(), ingresos.values(), color='mediumseagreen')
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso Total ($)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Imprimir ingresos en consola
for tienda, ingreso in ingresos.items():
    print(f"{tienda}: ${ingreso:,.0f}")
```

# 2. Ventas por categoría

```python
# Contar productos vendidos por categoría para cada tienda
cat_tienda1 = tienda['Categoría del Producto'].value_counts()
cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
cat_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Mostrar gráficamente las categorías más populares de cada tienda
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Productos Vendidos por Categoría en Cada Tienda", fontsize=16)

# Tienda 1
axs[0, 0].bar(cat_tienda1.index, cat_tienda1.values, color='cornflowerblue')
axs[0, 0].set_title("Tienda 1")
axs[0, 0].tick_params(axis='x', rotation=45)

# Tienda 2
axs[0, 1].bar(cat_tienda2.index, cat_tienda2.values, color='darkseagreen')
axs[0, 1].set_title("Tienda 2")
axs[0, 1].tick_params(axis='x', rotation=45)

# Tienda 3
axs[1, 0].bar(cat_tienda3.index, cat_tienda3.values, color='sandybrown')
axs[1, 0].set_title("Tienda 3")
axs[1, 0].tick_params(axis='x', rotation=45)

# Tienda 4
axs[1, 1].bar(cat_tienda4.index, cat_tienda4.values, color='lightcoral')
axs[1, 1].set_title("Tienda 4")
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular el promedio de calificaciones de los clientes por tienda
promedios = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

# Mostrar los promedios
for tienda, calificacion in promedios.items():
    print(f"{tienda}: {calificacion:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(promedios.keys(), promedios.values(), color='goldenrod')
plt.ylim(0, 5)
plt.ylabel("Calificación Promedio")
plt.title("Satisfacción del Cliente por Tienda")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Función para mostrar productos más y menos vendidos
def mostrar_top_productos(df, nombre_columna, nombre_tienda, color):
    conteo = df[nombre_columna].value_counts()
    top_5 = conteo.head(5)
    bottom_5 = conteo.tail(5)

    fig, ax = plt.subplots(1, 2, figsize=(14, 4))
    fig.suptitle(f"Productos más y menos vendidos - {nombre_tienda}", fontsize=14)

    # Más vendidos
    ax[0].bar(top_5.index, top_5.values, color=color)
    ax[0].set_title("Más vendidos")
    ax[0].tick_params(axis='x', rotation=45)

    # Menos vendidos
    ax[1].bar(bottom_5.index, bottom_5.values, color=color)
    ax[1].set_title("Menos vendidos")
    ax[1].tick_params(axis='x', rotation=45)

    plt.tight_layout(rect=[0, 0.03, 1, 0.90])
    plt.show()

# Nombre real de la columna de productos
nombre_col = [col for col in tienda.columns if "producto" in col.lower()][0]
print(f"Columna usada para productos: {nombre_col}")

# Mostrar para cada tienda
mostrar_top_productos(tienda, nombre_col, "Tienda 1", 'skyblue')
mostrar_top_productos(tienda2, nombre_col, "Tienda 2", 'lightgreen')
mostrar_top_productos(tienda3, nombre_col, "Tienda 3", 'salmon')
mostrar_top_productos(tienda4, nombre_col, "Tienda 4", 'plum')
```

# 5. Envío promedio por tienda

```python
# Detectar automáticamente la columna que contiene 'envío'
col_envio = [col for col in tienda.columns if "env" in col.lower()][0]
print(f"Columna detectada para envío: {col_envio}")

# Calcular promedio de envío por tienda
envios_promedio = {
    "Tienda 1": tienda[col_envio].mean(),
    "Tienda 2": tienda2[col_envio].mean(),
    "Tienda 3": tienda3[col_envio].mean(),
    "Tienda 4": tienda4[col_envio].mean()
}

# Mostrar resultados numéricos
for tienda_nombre, promedio in envios_promedio.items():
    print(f"{tienda_nombre}: ${promedio:,.2f}")

# Visualizar con gráfico de barras
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(envios_promedio.keys(), envios_promedio.values(), color='darkorange')
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo Promedio de Envío ($)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
